```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.shape
```

```python
df = df.astype(dtype={'id':str})
```

```python
df.info()
```

```python
df.describe()
```

```python
df = pd.get_dummies(df, columns=['Kelas Pekerja', 'Status Perkawinan','Pekerjaan','Jenis Kelamin'])
df.head()
```

```python
Pend = {'SD':0, '1st-4th':1, '5th-6th':2, '7th-8th':3, '9th':4, '10th':5, '11th':6, '12th':7, 'SMA':8, 'Pendidikan Tinggi':9, 'D4':10, 'D3':11,'Sarjana':12, 'Master':13, 'Sekolah Professional':14, 'Doktor':15}
Gj = {'<=7jt':0, '>7jt':1}
```

```python
df['Pendidikan'] = df['Pendidikan'].replace(Pend)
df['Gaji'] = df['Gaji'].replace(Gj)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdscalar = StandardScaler()
```

```python
df['Umur'] = StandardScaler().fit(df[['Umur']]).transform(df[['Umur']])
df['Berat Akhir'] = StandardScaler().fit(df[['Berat Akhir']]).transform(df[['Berat Akhir']])
df['Jmlh Tahun Pendidikan'] = StandardScaler().fit(df[['Jmlh Tahun Pendidikan']]).transform(df[['Jmlh Tahun Pendidikan']])
df['Keuntungan Kapital'] = StandardScaler().fit(df[['Keuntungan Kapital']]).transform(df[['Keuntungan Kapital']])
df['Kerugian Capital'] = StandardScaler().fit(df[['Kerugian Capital']]).transform(df[['Kerugian Capital']])
df['Jam per Minggu'] = StandardScaler().fit(df[['Jam per Minggu']]).transform(df[['Jam per Minggu']])
df.head()
```

```python
df.to_csv('data_train.csv', index=False)
```

```python
X_train = df.drop(['Gaji','id'], axis=1)
y_train = df['Gaji']
```

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
pipe = Pipeline(steps=[('estimator', SVC())])

# Add a dict of estimator and estimator related parameters in this list
params_grid = [{
                'estimator':[SVC()],
                'estimator__C': [1, 10, 100, 1000],
                'estimator__gamma': [0.001, 0.0001],
                }, 
                {
                'estimator':[knn()],
                'estimator__n_neighbors':np.arange(5,10),
                'estimator__weights':['distance','uniform']
                }, 
                {
                'estimator': [DecisionTreeClassifier()],
                'estimator__max_depth': [1,20],
                'estimator__max_features': [None, "auto", "sqrt", "log2"],
                },
               # {'estimator':[Any_other_estimator_you_want],
               #  'estimator__valid_param_of_your_estimator':[valid_values]

              ]

grid = GridSearchCV(pipe, params_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_ 
```

```python
grid.best_params_ 
```

```python
grid.best_score_ 
```
