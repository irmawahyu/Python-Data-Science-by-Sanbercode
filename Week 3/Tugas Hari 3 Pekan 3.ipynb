```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(22,9))
ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, sin, color='r', label='sinX')
ax[2].plot(x, cos, color='g', label='cosX')

ax[0].set_title('Grafik tanx', fontsize=14)
ax[1].set_title('Grafik sinx', fontsize=14)
ax[2].set_title('Grafik cosx', fontsize=14)
fig.legend(loc='upper center', fontsize=12, framealpha=0.3)
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
plt.figure(figsize=(12,8))
ax = plt.axes()
ax1 = plt.axes((0.03, 0.8, 0.25, 0.25))
ax2 = plt.axes((0.8, 0.8, 0.25, 0.25))
ax3 = plt.axes((0.03, 0.03, 0.25, 0.25))
ax4 = plt.axes((0.8, 0.03, 0.25, 0.25))

ax.plot(x,y)
ax1.plot(x,y5, color='y')
ax1.plot(x,y, color='b')
ax2.plot(x,y4, color='k')
ax2.plot(x,y, color='b')
ax3.plot(x,y2, color='r')
ax3.plot(x,y, color='b')
ax4.plot(x,y3, color='g')
ax4.plot(x,y, color='b')

ax.set_title('Grafik CosX', pad=100)
ax1.set_title('Grafik CosX^5')
ax2.set_title('Grafik CosX^4')
ax3.set_title('Grafik CosX^2')
ax4.set_title('Grafik CosX^3')

plt.show()
```
